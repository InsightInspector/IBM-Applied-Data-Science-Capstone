## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
For Example "Landing_Outcome"**

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
query = "SELECT DISTINCT(Launch_Site) FROM SPACEXTBL"
results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = ['Launch_site'])
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
query = "SELECT * FROM SPACEXTBL WHERE Launch_site LIKE 'CCA%' LIMIT 5"
results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = df.columns)
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
query = """SELECT SUM(PAYLOAD_MASS__KG_) 
           FROM SPACEXTBL 
           WHERE Customer = 'NASA (CRS)'
           GROUP BY Customer"""
results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = ['Total Payload Mass Carried by Boosters Launched by NASA (CRS)'])
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
query = """SELECT AVG(PAYLOAD_MASS__KG_) 
           FROM SPACEXTBL 
           WHERE Booster_Version = 'F9 v1.1'
           GROUP BY Booster_Version"""
results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = ['Average Payload Mass Carried by Boosters Version F9 v1.1'])
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
query = """SELECT MIN(Date) 
           FROM SPACEXTBL 
           WHERE [Landing _Outcome] = 'Success (ground pad)'"""
results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = ['First Successful Landing Outcome in Ground Pad'])
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
query = """SELECT DISTINCT(Booster_Version) 
           FROM SPACEXTBL 
           WHERE [Landing _Outcome] = 'Success (drone ship)'
           AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000"""
results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = ['Names of Boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000'])
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
query = """SELECT
            SUM(CASE WHEN Mission_Outcome LIKE 'Suc%' THEN 1 ELSE 0 END) AS Success,
            SUM(CASE WHEN Mission_Outcome LIKE 'Fai%' THEN 1 ELSE 0 END) AS Failure
            FROM SPACEXTBL"""
results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = ['Success', 'Failure'])
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
query = """SELECT DISTINCT(Booster_Version)
           FROM SPACEXTBL
           WHERE PAYLOAD_MASS__KG_ IN (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL)
            """

results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = ['Booster_Version'])
```

### Task 9

##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
query = """SELECT 
           substr(Date,4,2) AS "Month",
           substr(Date,7,4) AS "Year",
           Booster_Version,
           Launch_Site,
           [Landing _Outcome]
           FROM SPACEXTBL
           WHERE Year = '2015'
           AND [Landing _Outcome] = 'Failure (drone ship)'
"""

results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = ['Month','Year', 'Booster_Version', 'Launch_Site', 'Landing_Outcome'])
```

### Task 10

##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
query = """SELECT Date, COUNT([Landing _Outcome]) as Count
           FROM SPACEXTBL
           WHERE Date BETWEEN '04-06-2010' AND '20-03-2017'
           AND[Landing _Outcome] LIKE 'Succ%'
           GROUP BY Date
           ORDER BY Count ASC"""

results = cur.execute(query).fetchall()

pd.DataFrame(
    data = results,
    columns = ['Date', 'Count of Successful Landing Outcomes'])
```
